## Resolvé el siguiente desafío:

**EJERCICIO 10:** dado el archivo denominado [log_catedras](https://archivos.linti.unlp.edu.ar/index.php/s/ofgZh41dUYMYX6H) que contiene  el registro de  accesos al entorno catedras.linti, se desea:
1. Generar una función que retorne la actividad de un usuario dado como parámetro. Esta función recibe un parámetro  opcional que indica: "TODO" para retornar la información completa, "EXPLICACION",  solo los accesos relacionados a las explicaciones de práctica (chequear si en la columna "Contexto del evento" aparece el texto  "BigBlueButton: Sala de VC de explicaciones de práctica"), "TEORIA",  solo los accesos relacionados al material de clases la teoría (chequear si en la columna "Contexto del evento" aparece el texto "Página: Material de las clases"). Por defecto la función retorna toda la información. 
2. Escribir un programa que, utilizando la función anterior, muestre un listado similar al siguiente, donde sólo se muestra la información sobre fecha y hora de acceso, recurso accedido y dirección IP. Notar que la columna "Recurso accedido" no muestra todo el texto sino los primeros caracteres de modo que quede prolijo el listado.
 
```
Usuario: Hypno
------------------------------------------------
Dia/hora             Recurso accedido     Dir IP
------------------------------------------------
3/04/23, 09:22:22    BigBlueButton: Sala  190.19.24.34
3/04/23, 19:22:22    Tarea: Entrega 2     190.19.24.34
```

````python
#Solución
import csv
import os

def buscarUsuario(dicc, buscado, comoImprime="todo"):
    """Esta funcion recibe de manera obligatoria un diccionario con las actividades del usuario, un nombre de usuario buscado.

    {'Hora': '3/04/23, 09:37:32',
    'Nombre completo del usuario': 'Butterfree',
    'Usuario afectado': '-',
    'Contexto del evento': 'Curso: Seminario de  Lenguajes - Opción Python (2023)',
    'Nombre evento': 'Informe de registro visualizado',
    'Descripción': 'The user with id 12 viewed the log report for the course with id 1159.',
    'Dirección IP': '186.137.41.137'}

    De manera opcional se recibr que actividades se quieren imprimir segun su key "Contexto del evento":
        
    'todo': retorna una lista con todas las actividades del usuario donde
        {key = "Contexto del evento" value = *}

    'explicacion': retorna una lista con todas las actividades del usuario donde
        {key = "Contexto del evento" value = BigBlueButton: Sala de VC de explicaciones de práctica}

    'teoria': retorna una lista con todas las actividades del uisuario donde
        {key = "Contexto del evento" value = Página: Material de las clases}

    """

    apariciones = [linea for linea in dicc if linea["Nombre completo del usuario"] == buscado]
    match comoImprime : 
        case 'explicacion':
            return [dicc for dicc in apariciones if dicc["Contexto del evento"] == "BigBlueButton: Sala de VC de explicaciones de práctica"]
        case 'teoria':
            return [dicc for dicc in apariciones if dicc["Contexto del evento"] == "Página: Material de las clases"]
        case _ :
            return apariciones

ruta = os.path.realpath('.')
ruta_archivo = os.path.join(ruta, "log_catedras.csv")
with open(ruta_archivo,'r' , encoding="utf-8") as archivo:
    dictActividades = csv.DictReader (archivo, delimiter=",")
    print(next(dictActividades))
    
    formaImpresion = input("""Ingrese la informacion que desea listar:
                    'todo': listar todas las actividades del usuario
                    'explicacion': listar las actividades referidas a las clases de explicacion
                    'teoria': listar las actividades referidas a las clases de teoria""")
    
    buscado = input("Ingrese el nombre de usuario que desea buscar: ").lower().capitalize()
    
    actividadesBuscado = buscarUsuario(dictActividades, buscado, formaImpresion)

print(f"Usuario {buscado}")
print("-" * 90)
print("{:<22} {:<20} {:<15}".format("Dia/Hora", "Recurso Accedido", "Direccion Ip"))
print("-" * 90)
for actividad in actividadesBuscado:
    print("{:<22} {:<20} {:<15}".format(actividad["Hora"], actividad["Contexto del evento"][:16], actividad["Dirección IP"]))
    
````
