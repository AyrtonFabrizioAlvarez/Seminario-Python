1. Generar una función que retorna la actividad de un usuario dado como parámetro. Esta función recibe un parámetro  opcional que indica: "TODO" para retornar la información completa, "FACULTAD", solo los accesos realizados desde la facultad (dirección IP comienza con "163.10") o "EXTERIOR, solo los accesos realizados con una IP que no comienza con "163.10". Por defecto la función retorna solo los accesos desde la facultad.

Notar que la columna "Recurso accedido" no muestra todo el texto sino los primeros caracteres de modo que quede prolijo el listado.

````python
#Solución
import csv,os
def actividadUser(user, info = "FACULTAD"):
    #Siguiendo el ejemplo que esta en la teoría, no se si es lo correcto porque es muy dependiente la ruta
    ruta = os.path.realpath(".")
    ruta = os.path.join(ruta,"Downloads","log_catedras.csv")
    with open(ruta, encoding="utf8") as archivo_csv:
        csv_reader = csv.DictReader(archivo_csv,delimiter=",")
        print(type(csv_reader))
        """encabezado = next(csv_reader) #Se queda con la primer linea tambien que no deberia de ser así(y esto debería ser 
        lo mas correcto para csv) segun el ejemplo de la teoria, lo saco y dejo todo en data por esa razon"""               
        data = list(csv_reader)
    print('termine de leer el archivo')
       
    activ = []
    #Esto debería ser algo con filter y lambda pero no se bien como hacerlo
    for linea in data:
        if (linea["Nombre completo del usuario"] == user):
            if (info == "TODO"):
                activ.append(linea)
            elif (info == "FACULTAD"):
                if (linea["Dirección IP"].startswith("163.10")):
                    activ.append(linea) 
            elif(info == "EXTERIOR"):
                if not (linea["Dirección IP"].startswith("163.10")):
                    activ.append(linea)
            
    return activ
#Probando    
usuario = "Charmander"                      #EL USUARIO INGRRESADO NO TIENE EN CUENTA MINUSCULAS O MAYUSCULAS
act = actividadUser(usuario, 'EXTERIOR')    #EL PARAMETRO NO TIENE EN CUENTA MINUSCULAS O MAYUSCULAS
print(act)
````
